### `$match`
`$match` ë‹¨ê³„ëŠ” ì¡°ê±´ì— ë§ëŠ” ë„íë¨¼íŠ¸ë§Œ ë‹¤ìŒ ë‹¨ê³„ë¡œ ì „ë‹¬í•©ë‹ˆë‹¤.

$matchëŠ” MongoDBì˜ ë‹¤ì–‘í•œ ë¹„êµ ì—°ì‚°ìë¥¼ í•¨ê»˜ ì‚¬ìš©í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.

- `$eq`  : ê°™ë‹¤ (=)

- `$gte`: í¬ê±°ë‚˜ ê°™ë‹¤ (>=)

- `$lte`: ì‘ê±°ë‚˜ ê°™ë‹¤ (<=)

- `$gt` : í¬ë‹¤ (>)

- `$lt` : ì‘ë‹¤ (<)

- `$ne` : ê°™ì§€ ì•Šë‹¤ (!=)

- `$in` : ì—¬ëŸ¬ ê°’ ì¤‘ ì¼ì¹˜í•˜ëŠ” ê²ƒ ì¡°íšŒ

- `$nin` : ë°°ì—´ ì•ˆì— ì—†ëŠ” ê²½ìš°

```python
docs = movies.aggregate([
    {
        "$match": {
            # "year": {"$gte": 2001, "$lt": 2010}
            "rated": {"$in": ["TV-G", "G"]}
        }
    },
    {"$limit": 2}
])

for doc in docs:
    display(doc)
```

### `$group`

ìì£¼ ì‚¬ìš©í•˜ëŠ” ì§‘ê³„ ì—°ì‚°ì:
  - $sum : í•©ê³„ (count ìš©ë„ë¡œë„ ì‚¬ìš©)

  - $avg : í‰ê· 

  - $min : ìµœì†Œê°’

  - $max : ìµœëŒ€ê°’

ì£¼ì˜ì‚¬í•­:
  - _id ê°’ì„ nullë¡œ ì„¤ì •í•˜ë©´ ì „ì²´ ì§‘ê³„ê°€ ê°€ëŠ¥í•¨ (ex: ì „ì²´ í‰ê· )
  - $group ë‚´ë¶€ì—ì„œëŠ” ì›ë³¸ ë¬¸ì„œì˜ í•„ë“œë¥¼ ì§ì ‘ ì‚¬ìš©í•˜ë ¤ë©´ `$í•„ë“œëª…` ìœ¼ë¡œ ì°¸ì¡°í•´ì•¼ í•¨

```python
# ì—°ë„ë³„ ì˜í™” ê°œìˆ˜ êµ¬í•˜ê¸°: ì—°ë„ë³„ë¡œ ê·¸ë£¹í•‘ -> ê°ê° ê°œìˆ˜
docs = movies.aggregate([
    {
      "$group": {
          "_id": "$year",
          "movie_count": {"$sum": 1}
      }
    },
    {"$sort": {"_id": -1}}
])

for doc in docs:
  display(doc)
```

### `$project`

`$project`ëŠ” SQLì˜ SELECTì²˜ëŸ¼ ì›í•˜ëŠ” í•„ë“œë§Œ ë°˜í™˜í•˜ê±°ë‚˜ ìƒˆ í•„ë“œë¥¼ ê³„ì‚°í•˜ëŠ” ë‹¨ê³„ì…ë‹ˆë‹¤. 4ì¼ì°¨ì— ì§„í–‰í–ˆë˜ `projection`ê³¼ ë™ì¼í•œ ê°œë…ì…ë‹ˆë‹¤.

ê¸°ëŠ¥:
  - í•„ë“œ í¬í•¨/ì œì™¸ (1: í¬í•¨, 0: ì œì™¸)

  - í•„ë“œ ì´ë¦„ ë³€ê²½

  - ìƒˆë¡œìš´ ê³„ì‚° í•„ë“œ ìƒì„±(ì‚°ìˆ  ì—°ì‚° ê°€ëŠ¥)

  - ì¤‘ì²©ëœ í•„ë“œ ì ‘ê·¼ (ex. imdb.rating)

ì£¼ì˜:
  - `_id`ëŠ” ê¸°ë³¸ì ìœ¼ë¡œ í•­ìƒ í¬í•¨ë˜ë¯€ë¡œ ì œì™¸í•˜ë ¤ë©´ `_id: 0` í•´ì•¼ í•¨


```python
docs1 = movies.aggregate([
    {
        "$project": {
            "title": 1,
            "rating": "$imdb.rating",
            "_id": 0
        }
    }, {"$limit": 2}
])

docs2 = movies.aggregate([
    {
        "$project": {
            "title": 1,
            "score10": {"$multiply": ["$imdb.rating", 10]},
            "_id": 0
        }
    }, {"$limit": 2}
])

for doc in docs2:
    display(doc)
```

## ë¬¸ì œ 3) ì‹¬í˜ˆê´€ ìœ„í—˜ë„ê°€ ë†’ì€ ì‚¬ëŒ 3ëª… ì°¾ê¸°
- LDL ì ìˆ˜ = ldl ê°’ ê·¸ëŒ€ë¡œ
- í˜ˆì•• ì ìˆ˜ = ìˆ˜ì¶•ê¸° í˜ˆì•• sys
- í˜ˆë‹¹ ì ìˆ˜ = glucose

ê° í™˜ìì˜ ìœ„í—˜ë„(risk)ëŠ” ë‹¤ìŒì˜ í•©ìœ¼ë¡œ ê³„ì‚°í•©ë‹ˆë‹¤:
`risk = ldl + systolic + glucose`

Aggregationì„ ì‚¬ìš©í•˜ì—¬
1) í•„ìš”í•œ í•„ë“œë§Œ ì„ íƒí•˜ê³   
2) ìƒˆë¡œìš´ í•„ë“œ(risk)ë¥¼ ìƒì„±í•˜ì—¬ ê³„ì‚°í•œ ë’¤  
3) risk ê°’ì„ ê¸°ì¤€ìœ¼ë¡œ ë‚´ë¦¼ì°¨ìˆœ ì •ë ¬í•˜ì—¬  

ê°€ì¥ ìœ„í—˜ë„ê°€ ë†’ì€ 3ëª…ì„ ì¶œë ¥í•˜ì„¸ìš”.


```python
docs = col.aggregate([
    {
        "$project": {
            "name": 1,
            "ldl": 1,
            "systolic": 1,
            "glucose": 1,
            "_id": 0,
            "risk": {"$add": ["$ldl", "$systolic", "$glucose"]}
        }
    },
    {"$sort": {"risk": -1}},
    {"$limit": 3}
])

for doc in docs:
  display(doc)
```
