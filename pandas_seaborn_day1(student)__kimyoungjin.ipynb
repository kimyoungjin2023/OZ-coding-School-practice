## 2-3. Pandas ê¸°ë³¸ ì‚¬ìš©ë²• â€“ ì¡°ê±´, ì •ë ¬, ìˆ˜ì •, ì¶”ê°€/ì‚­ì œ

**ìì£¼ ì“°ëŠ” ê¸°ë³¸ ì¡°ì‘ë“¤**

- í–‰ ì¡°ê±´ í•„í„°ë§: `df[df["Score"] > 80]`  
- ì—¬ëŸ¬ ì¡°ê±´ ê²°í•©: `df[(df["Score"] > 80) & (df["Age"] < 35)]`

- ì»¬ëŸ¼ ì‚­ì œ: `df.drop(columns=["Passed"], inplace=True)`

### ğŸ”¹ ì‹¤ìŠµ ì½”ë“œ 2-3 â€“ Pandas ê¸°ë³¸ ì¡°ì‘ ì¢…í•© ì—°ìŠµ

````python
data = {
    "Name": ["Tom", "Jane", "Alice", "Bob"],
    "Age": [28, 34, 29, 42],
    "Score": [85, 92, 88, 75]
}
df_sample = pd.DataFrame(data)
df_sample.index = ["a", "b", "c", "d"]
print("=== ì›ë³¸ ë°ì´í„° ===")
display(df_sample)
````

````python

# 1) ì ìˆ˜ê°€ 80ì  ì´ˆê³¼ì¸ í•™ìƒ
print("=== Score > 80ì¸ í•™ìƒ ===")
display(df_sample[df_sample["Score"] > 80])
````

````python
# 2) Score > 80 AND Age < 35
print("=== Score > 80 AND Age < 35 ===")
display(df_sample[(df_sample["Score"] > 80) & (df_sample["Age"] < 35)])
````

````python
# 3) Score ê¸°ì¤€ ë‚´ë¦¼ì°¨ìˆœ ì •ë ¬
print("=== Score ê¸°ì¤€ ë‚´ë¦¼ì°¨ìˆœ ì •ë ¬ ===")
display(df_sample.sort_values(by="Score", ascending=False))
````

````python
# 4) Tomì˜ Scoreë¥¼ 90ì ìœ¼ë¡œ ìˆ˜ì •
df_sample.loc[df_sample["Name"] == "Tom", "Score"] = 90
print("=== Tomì˜ ì ìˆ˜ ìˆ˜ì • í›„ ===")
display(df_sample)
````

````python
# 5) Passed ì»¬ëŸ¼ ì¶”ê°€ (Score >= 80)
df_sample["Passed"] = df_sample["Score"] >= 80
print("=== Passed ì»¬ëŸ¼ ì¶”ê°€ í›„ ===")
display(df_sample)
````

````python
# 6) Passed ì»¬ëŸ¼ ì‚­ì œ
df_sample.drop(columns=["Passed"], inplace=True)
print("=== Passed ì»¬ëŸ¼ ì‚­ì œ í›„ ===")
display(df_sample)
````

````python
# 7) ì¸ë±ìŠ¤ 'b' í–‰ ì‚­ì œ
df_sample.drop(index="b", inplace=True)
print("=== ì¸ë±ìŠ¤ 'b' í–‰ ì‚­ì œ í›„ ===")
display(df_sample)
````
