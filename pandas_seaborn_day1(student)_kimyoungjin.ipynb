## 2-3. Pandas Í∏∞Î≥∏ ÏÇ¨Ïö©Î≤ï ‚Äì Ï°∞Í±¥, Ï†ïÎ†¨, ÏàòÏ†ï, Ï∂îÍ∞Ä/ÏÇ≠Ï†ú

**ÏûêÏ£º Ïì∞Îäî Í∏∞Î≥∏ Ï°∞ÏûëÎì§**

- Ìñâ Ï°∞Í±¥ ÌïÑÌÑ∞ÎßÅ: `df[df["Score"] > 80]`  
- Ïó¨Îü¨ Ï°∞Í±¥ Í≤∞Ìï©: `df[(df["Score"] > 80) & (df["Age"] < 35)]`  
- Ï†ïÎ†¨: `df.sort_values(by="Score", ascending=False)`  
- Í∞í ÏàòÏ†ï: `df.loc[df["Name"] == "Tom", "Score"] = 90`  
- ÏÉàÎ°úÏö¥ Ïª¨Îüº Ï∂îÍ∞Ä: `df["Passed"] = df["Score"] >= 80`  
- Ïª¨Îüº ÏÇ≠Ï†ú: `df.drop(columns=["Passed"], inplace=True)`  
- Ìñâ ÏÇ≠Ï†ú: `df.drop(index="b", inplace=True)`

### üîπ Ïã§Ïäµ ÏΩîÎìú 2-3 ‚Äì Pandas Í∏∞Î≥∏ Ï°∞Ïûë Ï¢ÖÌï© Ïó∞Ïäµ

```python
data = {
    "Name": ["Tom", "Jane", "Alice", "Bob"],
    "Age": [28, 34, 29, 42],
    "Score": [85, 92, 88, 75]
}
df_sample = pd.DataFrame(data)
df_sample.index = ["a", "b", "c", "d"]
print("=== ÏõêÎ≥∏ Îç∞Ïù¥ÌÑ∞ ===")
display(df_sample)
```

```python

# 1) Ï†êÏàòÍ∞Ä 80Ï†ê Ï¥àÍ≥ºÏù∏ ÌïôÏÉù
print("=== Score > 80Ïù∏ ÌïôÏÉù ===")
display(df_sample[df_sample["Score"] > 80])
```

```python
# 2) Score > 80 AND Age < 35
print("=== Score > 80 AND Age < 35 ===")
display(df_sample[(df_sample["Score"] > 80) & (df_sample["Age"] < 35)])
```

```python
# 3) Score Í∏∞Ï§Ä ÎÇ¥Î¶ºÏ∞®Ïàú Ï†ïÎ†¨
print("=== Score Í∏∞Ï§Ä ÎÇ¥Î¶ºÏ∞®Ïàú Ï†ïÎ†¨ ===")
display(df_sample.sort_values(by="Score", ascending=False))
```

```python
# 4) TomÏùò ScoreÎ•º 90Ï†êÏúºÎ°ú ÏàòÏ†ï
df_sample.loc[df_sample["Name"] == "Tom", "Score"] = 90
print("=== TomÏùò Ï†êÏàò ÏàòÏ†ï ÌõÑ ===")
display(df_sample)
```

```python
# 5) Passed Ïª¨Îüº Ï∂îÍ∞Ä (Score >= 80)
df_sample["Passed"] = df_sample["Score"] >= 80
print("=== Passed Ïª¨Îüº Ï∂îÍ∞Ä ÌõÑ ===")
display(df_sample)
```

```python
# 6) Passed Ïª¨Îüº ÏÇ≠Ï†ú
df_sample.drop(columns=["Passed"], inplace=True)
print("=== Passed Ïª¨Îüº ÏÇ≠Ï†ú ÌõÑ ===")
display(df_sample)
```

```python
# 7) Ïù∏Îç±Ïä§ 'b' Ìñâ ÏÇ≠Ï†ú
df_sample.drop(index="b", inplace=True)
print("=== Ïù∏Îç±Ïä§ 'b' Ìñâ ÏÇ≠Ï†ú ÌõÑ ===")
display(df_sample)
```

### üîπ ÎØ∏Îãà Î¨∏Ï†ú 2-3
1. `Score`Í∞Ä **90Ï†ê Ïù¥ÏÉÅ**Ïù∏ ÌïôÏÉùÎßå ÌïÑÌÑ∞ÎßÅÌïòÎäî ÏΩîÎìúÎ•º ÏßÅÏ†ë ÏûëÏÑ±Ìï¥ Î≥¥ÏÑ∏Ïöî.  
2. ÏÉà Ïª¨Îüº `Level`ÏùÑ ÎßåÎì§Ïñ¥, `Score`Í∞Ä 85Ï†ê Ïù¥ÏÉÅÏù¥Î©¥ `'High'`, Í∑∏Î†áÏßÄ ÏïäÏúºÎ©¥ `'Normal'`ÏùÑ ÎÑ£ÏúºÎ†§Î©¥  
   `df.loc[Ï°∞Í±¥, "Level"] = Í∞í` ÌòïÌÉúÎ•º ÌôúÏö©Ìï¥ Ïñ¥Îñ§ ÏΩîÎìúÎ•º ÏûëÏÑ±Ìï† Ïàò ÏûàÏùÑÍπåÏöî?


```python
# 1) ScoreÍ∞Ä 90Ï†ê Ïù¥ÏÉÅÏù∏ ÌïôÏÉù
display(df_sample[df_sample["Score"]>=90])
```

```python
# 2) Level Ïª¨Îüº Ï∂îÍ∞Ä
df_sample.loc[df_sample["Score"] >= 85, "Level"] = "High"
df_sample.loc[df_sample["Score"] < 85, "Level"] = "Normal"
df_sample
```

## 3. Ïã§Î¨¥Ìòï ÎØ∏ÏÖò ‚Äì ÎãπÎá®Î≥ë Îç∞Ïù¥ÌÑ∞Î°ú ÏóÖÎ¨¥ ÏöîÏ≤≠ Ìï¥Í≤∞ÌïòÍ∏∞

ÏïÑÎûò ÎØ∏ÏÖòÎì§ÏùÄ Ï†ÑÏ≤òÎ¶¨ ÏôÑÎ£åÎêú `df`Î•º ÏÇ¨Ïö©Ìï©ÎãàÎã§.

### [Í≥µÌÜµ Ï§ÄÎπÑ ÏÖÄ] ‚Äì Ï†ÑÏ≤òÎ¶¨ ÏôÑÎ£å DataFrame ÎßåÎì§Í∏∞

```python
import pandas as pd
import numpy as np
from IPython.display import display

df = pd.read_csv("train.csv")
potential_zeros = ["Glucose", "BloodPressure", "SkinThickness", "Insulin", "BMI"]
df[potential_zeros] = df[potential_zeros].replace(0, np.nan)
numeric_cols = df.select_dtypes(include=["int64", "float64"]).columns
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].mean())
print("Ï†ÑÏ≤òÎ¶¨ ÏôÑÎ£å! Í≤∞Ï∏°Ïπò Í∞úÏàò:")
print(df.isnull().sum())
display(df.head())
```

### ÎØ∏ÏÖò 1. ‚ÄúÌèâÍ∑† ÌòàÎãπ Î≥¥Í≥†ÏÑú Î∂ÄÌÉÅÎìúÎ†§Ïöî‚Äù

**ÏóÖÎ¨¥ ÏöîÏ≤≠**

> ‚ÄúÏ†ÑÏ≤¥ ÌôòÏûêÏùò ÌèâÍ∑† ÌòàÎãπÏù¥ Ïñ¥Îäê Ï†ïÎèÑÏù∏ÏßÄ ÏïåÍ≥† Ïã∂ÏäµÎãàÎã§.  
> Í∑∏Î¶¨Í≥† ÎãπÎá®Î≥ë ÌôòÏûê(Outcome=1)ÏôÄ ÎπÑÎãπÎá®Î≥ë ÌôòÏûê(Outcome=0)Ïùò ÌèâÍ∑† ÌòàÎãπ Ï∞®Ïù¥ÎèÑ Í∞ôÏù¥ ÏïåÎ†§Ï£ºÏÑ∏Ïöî.‚Äù


```python
overall_glucose_mean = df["Glucose"].mean()
grouped_glucose = df.groupby("Outcome")["Glucose"].mean()
print(f"Ï†ÑÏ≤¥ ÌèâÍ∑† ÌòàÎãπ(Glucose): {overall_glucose_mean:.2f}\n")
print("OutcomeÎ≥Ñ ÌèâÍ∑† ÌòàÎãπ:")
print(grouped_glucose)
```

### ÎØ∏ÏÖò 2. ‚ÄúÎãπÎá® ÌôòÏûê Ïó∞Î†πÎåÄ ÏöîÏïΩ Ï¢Ä Ìï¥Ï£ºÏÑ∏Ïöî‚Äù

**ÏóÖÎ¨¥ ÏöîÏ≤≠**

> ‚ÄúÎãπÎá® ÌôòÏûê(Outcome=1)Ïùò ÏµúÏÜå/ÏµúÎåÄ/ÌèâÍ∑†/Ï§ëÏïôÍ∞í ÎÇòÏù¥Î•º ÏïåÍ≥† Ïã∂ÏäµÎãàÎã§.‚Äù


```python
diabetic = df[df["Outcome"] == 1]
age_min = diabetic["Age"].min()
age_max = diabetic["Age"].max()
age_mean = diabetic["Age"].mean()
age_median = diabetic["Age"].median()
print("=== ÎãπÎá® ÌôòÏûê(Outcome=1) Ïó∞Î†π ÏöîÏïΩ ===")
print(f"ÏµúÏÜå ÎÇòÏù¥ : {age_min}")
print(f"ÏµúÎåÄ ÎÇòÏù¥ : {age_max}")
print(f"ÌèâÍ∑† ÎÇòÏù¥ : {age_mean:.2f}")
print(f"Ï§ëÏïôÍ∞í   : {age_median}")
```

### ÎØ∏ÏÖò 3. ‚ÄúBMI ÎÜíÏùÄ Ïàú TOP 10 ÌôòÏûê ÎΩëÏïÑÏ£ºÏÑ∏Ïöî‚Äù

**ÏóÖÎ¨¥ ÏöîÏ≤≠**

> ‚ÄúBMIÍ∞Ä ÎÜíÏùÄ ÌôòÏûê 10Î™ÖÏùÑ ID, Age, BMI, OutcomeÎßå Î≥¥Ïù¥ÎèÑÎ°ù ÎΩëÏïÑ Ï£ºÏÑ∏Ïöî.‚Äù


```python
top10_bmi = df.sort_values(by="BMI", ascending=False)[["ID", "Age", "BMI", "Outcome"]].head(10)
from IPython.display import display
display(top10_bmi)
```
